```python
url = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020'
    
```

```python
file_paths = [f'{url}-10.csv.gz', f'{url}-11.csv.gz', f'{url}-12.csv.gz']
```

```python
file_paths
```

```python
dfs = []
taxi_dtypes = {
                    'VendorID': pd.Int64Dtype(),
                    'store_and_fwd_flag':str,
                    'RatecodeID':pd.Int64Dtype(),
                    'PULocationID':pd.Int64Dtype(),
                    'DOLocationID':pd.Int64Dtype(),
                    'passenger_count': pd.Int64Dtype(),
                    'trip_distance': float,
                    'fare_amount': float,
                    'extra':float,
                    'mta_tax':float,
                    'tip_amount':float,
                    'tolls_amount':float,
                    'ehail_fee':float, 
                    'improvement_surcharge':float,
                    'total_amount':float,
                    'payment_type': pd.Int64Dtype(),
                    'trip_type': float,
                    'congestion_surcharge':float
                }
# native date parsing 
parse_dates = ['lpep_pickup_datetime', 'lpep_dropoff_datetime']
```

```python
for file_path in file_paths:
    # Read CSV file into a DataFrame and append to the list
        df = pd.read_csv(file_path, sep=',', compression='gzip', dtype=taxi_dtypes, parse_dates=parse_dates)
        dfs.append(df)
```

```python
len(dfs)
```

```python
final_df = pd.concat(dfs, ignore_index=True)
```
